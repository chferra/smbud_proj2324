```python
lang = pd.read_csv("https://raw.githubusercontent.com/datasets/language-codes/master/data/language-codes.csv")
lang.columns=['original_language','language']
lang.head(3)
```

```python
# uniamo i due dataset con un'operazione di join (metodo .merge() della libreria Pandas)
df = df.merge(lang, how='left', on='original_language')
df.head(3)
```

```python
#drop adult films
df_out = df.copy()
df_out = df_out[df_out['adult']==False]
df_out.shape
```

```python
df_out[df_out['release_date'].isna()].shape[0]
```

```python
# drop movies without release date
df_out.drop(df_out[df_out['release_date'].isna()].index, inplace=True)
```

```python
df_out[df_out['production_countries'].isna()].shape[0]
```

```python
#drop movies without production countries
df_out.drop(df_out[df_out['production_countries'].isna()].index, inplace=True)
```

```python
df_out[df_out['production_companies'].isna()].shape[0]
```

```python
#drop movies without production companies
df_out.drop(df_out[df_out['production_companies'].isna()].index, inplace=True)
df_out[df_out['production_companies'].isna()].shape[0]
```

```python
#drop movies without genres
df_out.drop(df_out[df_out['genres'].isna()].index, inplace=True)
df_out[df_out['genres'].isna()].shape[0]
```

```python
#drop movies without language
df_out.drop(df_out[df_out['language'].isna()].index, inplace=True)
df_out[df_out['language'].isna()].shape[0]
```

```python
df_out.groupby('runtime', as_index = False).size().sort_values('size', ascending=False).head(10)
```

```python
#keep only released movies
df_out = df_out[df_out['status']=='Released']
df_out.shape
```

```python
#keep only movies with at least one vote and runtime greater than 0
df_out = df_out[df_out['vote_count'] > 0]
df_out = df_out[df_out['runtime'] > 0]
df_out.shape
```

```python
df_out.columns
```

```python
#Drop info about status(That is only released), adult(That are only not for adult), backdrop_path, homepage, imdb_id, original_language, original_title, poster_path, production_countries and spoken_languages
df_out = df_out[['id', 'title', 'vote_average', 'vote_count',
                 'release_date', 'revenue', 'runtime',
                 'budget', 'overview', 'popularity',
                 'tagline', 'genres', 'production_companies', 'language']]
print(f'Shape: {df_out.shape}')
df_out.head(3)
```
